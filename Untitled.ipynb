```python
import numpy as np
np.random.seed(42)
import tensorflow as tf
import random as rn
import os

from keras import applications
from jons_image import ImageDataGenerator
from keras import optimizers
from keras.models import Sequential, Model
from keras.layers import Input
from keras.layers.core import Dense, Dropout, Activation, Flatten, Lambda
from keras.layers.convolutional import MaxPooling2D, ZeroPadding2D, AveragePooling2D, Conv2D
from keras.optimizers import SGD, rmsprop
from keras.regularizers import l2
from keras.callbacks import ModelCheckpoint, TensorBoard, LearningRateScheduler
from keras import backend as K

import matplotlib.image as mpimg
from PIL import Image as pil_image
import rasterio
import time

now = time.strftime("%d%m%Y-%H%M")

def filter(iter):
    for batch in iter:
        # print(batch[0][:,:,:,0].flatten().mean(), batch[0][:,:,:,1].flatten().mean(), batch[0][:,:,:,2].flatten().mean(), batch[0][:,:,:,3].flatten().mean())
        yield batch[0] - means, batch[1]

# dimensions of our images.
img_width, img_height = 200, 200

train_data_dir = 'data/train/'
validation_data_dir = 'data/valid/'

nb_train_samples = 348
nb_validation_samples = 174
epochs = 10
batch_size = 32
bands = 3
num_classes = 2

if bands == 1:
    color_mode = "grayscale"
elif bands == 3:
    color_mode = "rgb"
elif bands > 3:
    color_mode = "multispectral"
print('process running',nb_train_samples,'training and',nb_validation_samples,'validation samples in',color_mode)

if K.image_data_format() == 'channels_first':
    input_shape = (bands, img_width, img_height)
else:
    input_shape = (img_width, img_height, bands)

# the augmentation was done directly from the big tif to include as much data
train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest')
test_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest')

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    color_mode=color_mode,
    nbands=bands,
    seed=1)

validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    color_mode=color_mode,
    nbands=bands,
    seed=0)

#mean of the train dataset
mean = np.zeros((1, bands, img_width, img_height))
N = 0
for subdir, dirs, files in os.walk(train_data_dir):
    for fName in files:
        
        with rasterio.open(os.path.join(subdir, fName)) as r:
            x = r.read()

        target_size = img_width, img_height
        if target_size != x[0].shape:
            oldx=x
            x=np.zeros((oldx.shape[0],target_size[0], target_size[1]))
            hw_tuple = (target_size[1], target_size[0])
            # the following is an ugly hack & is liable to break if pil doesn't support the array dtype
            for i in range(0,oldx.shape[0]):
                img = pil_image.fromarray(oldx[i])
                img = img.resize(hw_tuple)
                x[i]=np.asarray(img, dtype=K.floatx())
        elif x.dtype != K.floatx(): # need to ensure type is correct
            x=x.astype(dtype=K.floatx())

        if K.image_data_format() == 'channels_last':
            x = np.rollaxis(x,0,3)

        if x.shape == (img_width, img_height, 5):
            mean[0][0] += x[:, :, 0]
            mean[0][1] += x[:, :, 1]
            mean[0][2] += x[:, :, 2]
            mean[0][3] += x[:, :, 3]
            mean[0][4] += x[:, :, 4]
            N += 1
        if x.shape == (img_width, img_height, 4):
            mean[0][0] += x[:, :, 0]
            mean[0][1] += x[:, :, 1]
            mean[0][2] += x[:, :, 2]
            mean[0][3] += x[:, :, 3]
            N += 1
        if x.shape == (img_width, img_height, 3):
            mean[0][0] += x[:, :, 0]
            mean[0][1] += x[:, :, 1]
            mean[0][2] += x[:, :, 2]
            N += 1
        if x.shape == (img_width, img_height, 1):
            mean[0][0] += x[:, :, 0]
            N += 1
mean[0] /= N
means = np.transpose(mean[0].astype(np.uint8), (1, 2, 0))[0][0]

print('images seen for calculating mean', N)
print('means: ',means)


def larger_model(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(32, (3, 3), input_shape=input_shape))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(128, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())
    model.add(Dense(64))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1))
    model.add(Activation('sigmoid', name='preds')) # because we have a 2 class problem (with more you'd want to use softmax)
    return model

# build the model
model = larger_model(train_generator.image_shape, num_classes)
model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)
```
